```python
import pandas as pd

df = pd.read_excel("data_science_project.xlsx")

print("Full Data:")
print(df)

print("\nFirst 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())

print("\nInfo:")
print(df.info())

print("\nMean:")
print(df.mean(numeric_only=True))

print("\nMedian:")
print(df.median(numeric_only=True))

print("\nMinimum values:")
print(df.min())

print("\nMaximum values:")
print(df.max())

print("\nCount:")
print(df.count())

print("\nMarks column:")
print(df['Marks'])

filtered = df[df['Marks'] > 80]

print("\nFiltered Data (Marks > 80):")
print(filtered)

sorted_df = df.sort_values(by='Marks', ascending=False)

print("\nSorted Data:")
print(sorted_df)

filtered.to_excel("filtered_data.xlsx", index=False)

print("\nFiltered data saved as 'filtered_data.xlsx'")
```
